```python
import requests
r = requests.get('https://awoiaf.westeros.org/index.php/List_of_characters')
```

```python
print(r.status_code, r.headers, r.text)
```

```python
from bs4 import BeautifulSoup 
soup = BeautifulSoup(r.text,"lxml")
```

```python
links = soup.find_all("a")
pages = []
for link in links:
    s1 = set(["title", "href"])
    s2 = set(link.attrs.keys())
    if s1 == s2:
        pages.append((link['href'],link['href'].split('/')[-1]))  #"<a href = >"text</a>
pages = pages[:-5]
pages = set(pages)
#print(pages[10])
```

```python
for href, name in pages:
    if name == 'Aegon_V_Targarye':
        print(href)
        r = requests.get('http://awoiaf.westeros.org' + href)
        with open(name + '.html', 'w+', encoding = 'utf-8') as fp:
            fp.write(r.text)
            
```

```python
with open('Aegon_V_Targaryen.html', 'r', encoding = 'utf-8') as fp:
    html = fp.read()
soup = BeautifulSoup(html, 'lxml')
```

```python
soup.find('table', class_="infobox").b.text
```

```python
import re
tbody = soup.find('table', class_="infobox").tbody
rows = tbody.find_all("tr")
for row in rows:
    for child in row.children:
        if child.name == "th":
            if row.th.text == 'Full Name':
                print("Full name :",row.td.text)
            if row.th.text == 'Alias':
                alias = row.td.text
                regex = r"([a-z])([A-Z])"

                test_str = alias

                subst = "\\1 , \\2"

                # You can manually specify the number of replacements by changing the 4th argument
                result = re.sub(regex, subst, test_str, 0, re.MULTILINE)
                print("Alias :",result)
                print(result.split(','))
                print([t.strip() for t in result.split(',')]) #strip()去除首尾指定字符  t.strip( )去除首尾空格
```
