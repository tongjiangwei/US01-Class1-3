```python
import requests
test_country = list(countries.keys())[0]
url = "https://restcountries.eu/rest/v2/name/" + test_country
#print(test_country)
r = requests.get(url)
data = r .json()
print(test_country,data[0]['alpha2Code'])
```

```python
import requests
from tqdm import tnrange,tqdm_notebook
from time import sleep

for i in tnrange(len(countries.keys())):
    try:
        country = list(countries.keys())[i]
        url = "https://restcountries.eu/rest/v2/name/" + country
        #print(test_country)
        r = requests.get(url)
        data = r .json()
        countries[country]['alpha2code']=data[0]['alpha2Code']
    except KeyError:
        print(data)
        sleep(5)
```

```python
import requests
#from tqdm import tnrange,tqdm_notebook
#from time import sleep

for country in countries.keys():
    try:
        #country = list(countries.keys())[i]
        url = "https://restcountries.eu/rest/v2/name/" + country
        #print(test_country)
        r = requests.get(url)
        data = r .json()
        countries[country]['alpha2code']=data[0]['alpha2Code']
    except KeyError:
        print(data)
        sleep(5)
```

```python
for k,v in countries.items():
    country = ""
    if len(v.keys()) == 0:
        if 'not bri' in k.lower():
            country = k.split("(")[0].strip()
        elif k == "Thailiand":
            country = "thailand"
        else:
            country = k[4:]
        try:
            url = "https://restcountries.eu/rest/v2/name/" + country
            r = requests.get(url)
            data = r .json()
            countries[k]['alpha2code']=data[0]['alpha2Code']
        except KeyError:
            print(data)
            sleep(5)
                
```

```python
for k,v in countries.items():
    try:
        url = "http://api.worldbank.org/v2/countries/"+v["alpha2code"]+"/indicators/NY.GDP.MKTP.CD?format=json"
        r = requests.get(url)
        data = r.json()
        countries[k]["gdp"] =  float(data[1][0]["value"])
        sleep(2)
    except IndexError: 
        print(k, "has a problem",data)
    except TypeError:
        print(k, "has a problem",data)
```
